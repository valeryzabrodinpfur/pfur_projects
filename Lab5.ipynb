```python
import pandas as pd

# Загрузка данных из CSV-файла
df = pd.read_csv('students.csv')

# Вывод первых 5 строк
print("Первые 5 строк:")
print(df.head())

# Информация о данных
print("\nИнформация о данных:")
print(df.info())

# Статистическая сводка по числовым столбцам
print("\nСтатистика по данным:")
print(df.describe())

# Средний балл студентов
average_score = df['Score'].mean()
print(f"\nСредний балл студентов: {average_score:.2f}")

# Количество студентов в каждой группе
group_counts = df['Group'].value_counts()
print("\nКоличество студентов в каждой группе:")
print(group_counts)
```

```python
import pandas as pd

# Загрузка данных
df = pd.read_csv('students.csv')

# Фильтрация студентов с баллом выше 80
high_scorers = df[df['Score'] > 80]

# Сортировка по убыванию балла
high_scorers_sorted = high_scorers.sort_values(by='Score', ascending=False)
print("Студенты с баллом выше 80, отсортированные по убыванию балла:")
print(high_scorers_sorted)

# Поиск самого старшего студента
oldest_student = df.loc[df['Age'].idxmax()]
print("\nСамый старший студент:")
print(oldest_student)

# Поиск самого младшего студента
youngest_student = df.loc[df['Age'].idxmin()]
print("\nСамый младший студент:")
print(youngest_student)
```

```python
import pandas as pd

# Загрузка данных
df = pd.read_csv('students.csv')

# Проверка наличия пропусков в данных
print("Количество пропусков в каждом столбце:")
print(df.isnull().sum())

# Заполнение пропусков в столбце 'Score' средним значением
mean_score = df['Score'].mean()
df['Score'].fillna(mean_score, inplace=True)

# Удаление строк с пропусками в столбце 'Group'
df = df.dropna(subset=['Group'])

# Проверка после обработки
print("\nПосле обработки пропусков:")
print(df.isnull().sum())
```

```python
import pandas as pd

# Загрузка данных
df = pd.read_csv('students.csv')

# Группировка по столбцу 'Group' с вычислением среднего балла и медианного возраста
grouped = df.groupby('Group').agg({
    'Score': 'mean',
    'Age': 'median'
}).rename(columns={'Score': 'Average_Score', 'Age': 'Median_Age'})

print("Средний балл и медианный возраст по группам:")
print(grouped)

# Добавление столбца 'Passed' (1, если Score >= 60, иначе 0)
df['Passed'] = (df['Score'] >= 60).astype(int)

print("\nДанные с добавленным столбцом 'Passed':")
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv('students.csv')

# Построение гистограммы распределения баллов
plt.figure(figsize=(8, 5))
df['Score'].hist(bins=10, edgecolor='black')
plt.title('Гистограмма распределения баллов')
plt.xlabel('Баллы')
plt.ylabel('Количество студентов')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Построение столбчатой диаграммы среднего балла по группам
average_scores = df.groupby('Group')['Score'].mean()

plt.figure(figsize=(8, 5))
average_scores.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Средний балл по группам')
plt.xlabel('Группа')
plt.ylabel('Средний балл')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
